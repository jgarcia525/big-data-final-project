```python
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords

from string import digits

import pandas as pd
import string
from pandas import DataFrame

def pre_processor(df):  
    
    for i in range (1, df.shape[0]):
        
        comment = df.iloc[i][6]
        df.iat[i, 6] =  text_process(comment)
   
    return df

def text_process(tweet):
    
    # Remove html tags
    soup = BeautifulSoup(tweet, 'html5lib')
    tweet = soup.get_text()
    
    # Remove links
    tweet = re.sub(r"http\S+", "", tweet)
    
    # Remove usernames
    tweet = re.sub('@[^\s]+','', tweet)
    
    # Remove numbers 
    remove_digits = str.maketrans('', '', digits)
    tweet = tweet.translate(remove_digits)
    
    # Check characters to see if they are in punctuation
    nopunc = [char for char in tweet if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return ' '.join([word for word in nopunc.split() if word.lower() not in stopwords.words('english')])



# ---------------- Shuffle data  ----------------

# df = pd.read_csv('twitter-data-all.csv', header=None, encoding="latin")
# df_shuffled = df.sample(frac=1)
# df_shuffled.to_csv('twitter-data-shuffled.csv')
# ds = pd.read_csv('twitter-data-shuffled.csv', header=None, encoding="latin")

# ---------------- Pre-process data  ----------------
# df_clean = pre_processor(ds)
# df_clean.to_csv('twitter-data-clean.csv', header=None, encoding="latin")
# print ("Cleaning == complete")
# print (df_clean.head())


ds = pd.read_csv('twitter-data-clean.csv', header=None, encoding="latin")

# ---------------- Split data into train, test, and eval ----------------
# Eval file too big for github, so it is split in 3

df_train = ds.iloc[:300000]
df_test = ds.iloc[300000:600000]
df_eval1 = ds.iloc[600000:933333]
df_eval2 = ds.iloc[933333:1266666]
df_eval3 = ds.iloc[1266666:1600000]


# ---------------- Checking correct size of data frames ----------------

# train_s = df_train.shape
# test_s = df_test.shape
# eval1_s = df_eval1.shape
# eval2_s = df_eval2.shape
# eval3_s = df_eval3.shape
# print ("train size: ", train_s)
# print ("test size: ", test_s)
# print ("eval1 size: ", eval1_s)
# print ("eval2 size: ", eval2_s)
# print ("eval3 size: ", eval3_s)


# ---------------- Create csv files with pre-processed data ----------------

df_train.to_csv('training-clean.csv')
df_test.to_csv('testing-clean.csv')
df_eval1.to_csv('evaluation1-clean.csv')
df_eval2.to_csv('evaluation2-clean.csv')
df_eval3.to_csv('evaluation3-clean.csv')
print ("CSV Files Creation: Success!")
```
